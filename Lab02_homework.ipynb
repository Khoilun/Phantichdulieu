```python
import pandas as pd
import numpy as np

df = pd.read_csv("https://raw.githubusercontent.com/huynhhoc/phantichdulieu/main/Data/dulieuxettuyendaihoc.csv")
print(df.head(5))
```

**1.Hãy sắp xếp dữ liệu điểm DH1 theo thứ tự tăng dần**

```python
df.sort_values(by="DH1")
```

**2.Hãy sắp xếp dữ liệu điểm DH2 tăng dần theo nhóm giới tính**

```python
df_male = df[df['GT'] == 'M']
df_male

df_female = df[df['GT'] == 'F']
df_female

sorted_value_male = df_male.sort_values(by ='DH2')
sorted_value_female = df_female.sort_values(by = 'DH2')

print('Male')
display(sorted_value_male)
print('Female')
display(sorted_value_female)
```

**3.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT**

```python
print(df.groupby("KT")["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]))
df_KT=df.groupby("KT")["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]) 
print(pd.pivot_table(df_KT,columns="KT"))
```

**4.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT và KV**

```python
df_KT_KV = df.groupby(["KT","KV"])["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"]) 
print( pd.pivot_table(df_KT_KV,columns=["KT","KV"]))
```

**5.Hãy tạo pivot-table để thống kê các giá trị count, sum, mean, median, min, max, std, Q1, Q2 và Q3 của DH1 theo KT, KV và DT**

```python
df_KT_KV_DT = df.groupby(["KT","KV","DT"])["DH1"].agg(["count", "sum", "mean", "median", "min", "max","std"])  
print(pd.pivot_table(df_KT_KV_DT,columns=["KT","KV","DT"]))
```
